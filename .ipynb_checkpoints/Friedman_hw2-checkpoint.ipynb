## Problem 1

```python
# Consider the normal means problem. Plot the risk of the James Stein estimator
from scipy.stats import poisson

n = [10,80]  # n = [10, 80]

# hierarchical composotion of ||Y||^(-2)
def r(mu_norm, n=10):
    # calculate expectation from Poisson(xi/2)
    xi = mu_norm**2
    def f(N):
        return 1.0/(n-2+2*N)
    expct = poisson.expect(f, args=[xi/2])
    return n - (n-2)**2 * expct
```

```python
mu = np.linspace(0,50,num=100)

r_val = np.zeros((mu.size, len(n)))
for j in range(len(n)):    
    for i in range(mu.size):
        r_val[i,j] = r(mu[i],n[j])
r_val.shape
```

```python
fig = plt.figure()
plt.plot(mu, r_val[:,0], label='n=10')
plt.plot(mu, r_val[:,1], label='n=80')
plt.xlabel(r'$\Vert \mu \Vert$')
plt.ylabel(r'risk')
oldParams = plt.rcParams
plt.rcParams['figure.figsize'] = [10,5]
plt.legend()
plt.show()
plt.rcParams = oldParams
```

```python
# Calculate upper bound 
def r_ubound(mu_norm, n):
    return 2 + (n-2)*mu_norm**2/((n-2)+mu_norm**2)

r_upper = np.zeros((mu.size, len(n)))
for j in range(len(n)):    
    for i in range(mu.size):
        r_upper[i,j] = r_ubound(mu[i],n[j])
r_upper.shape
```

```python
plt.plot(mu, r_val[:,0], label='n=10')
plt.plot(mu, r_upper[:,0], label='Bound, n=10')
plt.xlabel(r'$\Vert \mu \Vert$')
plt.ylabel(r'risk')
oldParams = plt.rcParams
plt.rcParams['figure.figsize'] = [12,5]
plt.legend()
plt.show()
plt.rcParams = oldParams
```

```python
plt.plot(mu, r_val[:,1], label='n=80')
plt.plot(mu, r_upper[:,1], label='Bound, n=80')
plt.xlabel(r'$\Vert \mu \Vert$')
plt.ylabel(r'risk')
oldParams = plt.rcParams
plt.rcParams['figure.figsize'] = [12,5]
plt.legend()
plt.show()
plt.rcParams = oldParams
```

```python
# Load baseball data from .csv
df = pd.read_csv('baseball.csv')
df
```

```python
# let p denote the season average for each player
p = df['SeasonAverage']
p
```

```python
n = df['At-Bats'][0]      # number of At-Bats, n=45
h = df['Hits']            # number of hits with 45 at-bats
z = df['BattingAverage']  # batting average after 45 at-bats
```

Use the variance stabilizing transformation 
\begin{aligned}
y_i &= \sqrt{n}\arcsin{(2z_i-1)}    \\
\mu_i &= \sqrt{n}\arcsin{(2p_i-1)}
\end{aligned}

```python
y  = np.sqrt(n)*np.arcsin(2*z - 1)
mu = np.sqrt(n)*np.arcsin(2*p - 1)
```

Let the estimator for $\mu$ be the maximum likelihood estimator $\hat{\mu}^{\mathrm{ML}} = y_i$

Calculate the mean squared error $\sum_{i=1}^{18} (\hat{\mu}_i^\mathrm{ML} - \mu_i)^2$

```python
mu.name = 'mu'
y.name = 'mu_ML'
pd.concat([mu, y],axis=1)
```

```python
mse = np.sum((y-mu)**2)  # Calculate the mean squared error
mse
```

Let the estimator for $\mu$ be a modified James Stein estimator $\hat{\mu}^{\mathrm{JS}} = \bar{y} + \big(1 - \frac{N-3}{S}\big)(y_i-\bar{y})$

Calculate the mean squared error $\sum_{i=1}^{18} (\hat{\mu}_i^\mathrm{JS} - \mu_i)^2$

```python
ybar = np.mean(y)  # mean of y_i's
N = n
S = np.sum((y-ybar)**2)
mu_JS = ybar + (1 - (N-3)/S) * (y - ybar)
mu.name, mu_JS.name = 'mu', 'mu_JS'
display(pd.concat([mu, mu_JS],axis=1))
```

```python
mse = np.sum((mu_JS-mu)**2)  # Calculate the mean squared error for mu_JS
print('MSE = {}'.format(mse))
```

Since our model for hits is $h_i \sim \mathrm{Binomial}(45, p_i)$, use a Beta prior on $p_i$'s to get the posterior distribution

$$
p_i \mid y_i, a, b \sim \mathrm{Beta}(h_i + a, 45-h_i+b)
$$

Then the expectation of the posterior is 

$$
\mathrm{E}(p_i \mid y_i, a, b) = \frac{h_i + a}{45+a+b}
$$

Use the expectation to estimate $\hat{p}_i = \mathrm{E}(p_i \mid y_i, a, b)$. To estimate the values of $a, b$, use the method of moments estimator by assuming $\hat{p}_i = h_i/45$. We already have the mean of the Beta posterior. The variance is

$$
\mathrm{Var}(p_i \mid y_i, a, b) = \frac{ab}{(a+b)^2 (a+b+1)}
$$

```python
# Caculate var(p) from the data
var_p = np.var(df['BattingAverage'])
# var_p = np.var(p)
var_p
```

```python
# Solve set of equations for a, b
from scipy.optimize import root
```

```python
# quick and dirty m.l.e. estimator for E(p)
p_ML = df['BattingAverage']  # batting average at 45 at-bats

# create anonymous functions for method of moments
def method_moments(x, i):
    # x = [a,b]
    y = np.zeros(2)  # y = [mu, var]
    h = df['Hits']
    # posterior parameters
    #   alpha = h_i + a
    #   beta  = 45 - h_i + b
    a_pos = h[i]+x[0]
    b_pos = 45-h[i]+x[1]
    y[0] = (h[i]+x[0])/(45+x[0]+x[1]) - np.mean(p_ML)
    y[1] = (a_pos*b_pos)/((a_pos+b_pos)**2 * (a_pos+b_pos+1)) - var_p
    return y
```

```python
m = 18
a = np.zeros(m)
b = np.zeros(m)
```

```python
for i in range(m):
    def f(x):
        return method_moments(x, i)
    x0 = np.array([1.0, 1.0])
    soln = root(f, x0)
    a[i], b[i] = soln.x[0], soln.x[1]
```

```python
# Now that we have solutions for a, b we can calculate the bayesian estimator, p_BE
p_BE = 45.0/(45+a+b)*(h/45) + (a+b)/(45+a+b)*(a/(a+b))
```

```python
# Calculate the mean squared error for p_BE and p_ML
mse_BE = np.sum((p_BE-p)**2)
mse_ML = np.sum((p_ML-p)**2)
print('mse_BE = {:.10f}'.format(mse_BE))
print('mse_ML = {:.10f}'.format(mse_ML))

# Find percent gain of bayesian estimator over m.l.e.
```

## Problem 6

```python
from numpy.random import random
n = 100**2
s_list = [5, 20, 100, 500, 1000]

mu = np.zeros((len(s_list),n))
mu_tmp = np.zeros(n)
# draw s samples from uniform distribution for mean vector mu
for j in range(len(s_list)):
    s = s_list[j]
    mu_tmp[:s] = random(s)
    for i in range(s):
        if mu_tmp[i] >= 0.5:
            mu[j,i] = 1 + (mu_tmp[i]-0.5)/0.5 * 4
        else:
            mu[j,i] = -1 - (mu_tmp[i])/0.5 * 4
```

```python
# Draw independent samples from normal distribution with given mean mu
from scipy.stats import norm
x = np.zeros((len(s_list),n))
for j in range(len(s_list)):
    for i in range(n):
        x[j,i] = norm.rvs(loc=mu[j,i])
```

Consider the hard threshold estimator $\hat{\mu}_\lambda^{\mathrm{HT}}$ and define $r(\lambda)=\frac{1}{n}\Vert \hat{\mu}_\lambda^{\mathrm{HT}} - \mu \Vert^2$.

Make a plot of $r(\lambda)$ vs. $\lambda$ with $\lambda \in [0, \sqrt{2 \log n}]$

```python
# Set lambda values within bounds
lam = np.linspace(0, np.sqrt(2*np.log(n)),num=200)

# Calculate mu_HT
mu_HT = np.zeros((len(s_list, n, lam.size)))
for j in range(len(s_list)):
    for i in range(lam.size):
        idx = np.where(np.abs(x[j])>lam[i])[0]
        mu_HT[j,idx,i] = x[j,idx]

    # Calculate r(lam)
    r = np.zeros(lam.size)
    for i in range(lam.size):
        r[i] = (1.0/n) * np.sum((mu_HT[j,:,i]-mu)**2)
    # Plot results
    plt.plot(lam, r)
    plt.xlabel(r'$\lambda$', size=14)
    plt.ylabel(r'$r(\lambda)$', size=14)
```

```python
# Plot results
plt.plot(lam, r)
plt.xlabel(r'$\lambda$', size=14)
plt.ylabel(r'$r(\lambda)$', size=14)
plt.show()
```

## Problem 7

```python
n = 80
p = 50
lam = np.linspace(0, 2, num=201)  # vary lambda from [0,2] with spacing 0.01
```

```python
# Y (n,1) vector in hw3_y.txt
# X (n,p) matrix in hw3_x.txt
# true regression parameters in hw3_truebeta.txt

# Load data from text files
y = np.loadtxt('hw3_y.txt')
x = np.genfromtxt('hw3_x.txt',delimiter=',')
beta_true = np.loadtxt('hw3_truebeta.txt',)
```

```python
beta0=np.ones(p)
tmp=x.dot(beta0)
tmp.shape
```

```python
# set up ridge regression
from scipy.optimize import minimize

i = 50  # pick one for now
beta_approx = []
for i in range(lam.size):
    # create anonymous function for ridge regression minimization
    def q(beta):
        return np.sum((y-x.dot(beta))**2) + lam[i]*np.sum(beta**2)
    # minimize q with respect to beta
    beta0 = np.zeros(p)
    beta_approx.append(minimize(q, beta0).x)
```

```python
# Calculate the mean squared prediction error (MSPE)
mspe = np.zeros(len(beta_approx))
for i in range(lam.size):
    mspe[i] = np.sum((x.dot(beta_approx[i]) - x.dot(beta_true))**2)
```

```python
# Plot the MSPE with respect to lambda
plt.plot(lam, mspe)
plt.xlabel(r'$\lambda$')
plt.ylabel('MSPE')
plt.show()
```

```python
# Part (ii)
# Fix lambda = 0.4

# Get the ridge estimator
idx = int(np.where(lam==0.4)[0])  # index where lambda=0.4 from earlier minimization
beta_R = beta_approx[idx]
xbetaR = x.dot(beta_R)

# Get the OLS estimator
beta_OLS = beta_approx[0]  # index where lambda=0.0 from earlier minimization
xbetaOLS = x.dot(beta_OLS)

# Get mu
mu = x.dot(beta_true)

# Plot values
plt.plot(mu, xbetaR, '.', markersize=10, label=r'$X\hat{\beta}_R$')
plt.plot(mu, xbetaOLS, '.', markersize=10, label=r'$X\hat{\beta}_{OLS}$')
plt.xlabel(r'$\mu=X\beta_0$')
plt.ylabel(r'$X\hat{\beta}$')
plt.legend(fontsize=16)
plt.show()
```
